```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_linnerud
from sklearn.neighbors import LocalOutlierFactor
```

```python
df = load_linnerud(as_frame = True).frame
df.head()
```

```python
X = df[['Chins','Situps','Jumps']]
y = df[['Weight','Waist','Pulse']]
```

# Local Outlier Factor (LOF) algorithm

```python
lof = LocalOutlierFactor(n_neighbors=5)
```

```python
lof.fit(X)
```

```python
outlier_labels = lof.fit_predict(y)
```

```python
y['Outlier'] = outlier_labels
```

```python
y['Outlier'] = y['Outlier'].apply(lambda x: 'Outlier' if x == -1 else 'Normal')
```
